3. Avg Time for Trip Based on Starting Point

East 5th/Shady Ln is the longest trip at 197.470588 minutes, or 3.29 hours. The station with the least time comes in at 9.75 minutes.

```python
%%bigquery --project dataviz-488717
SELECT start_station_name,
    AVG(duration_minutes)
    AS avg_trip_minutes
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
ORDER BY avg_trip_minutes DESC;
```

4. How Many Trips Start at Each Starting Point

The most trips taken are at 108,559. The one right after is 71,145. There are pretty some pretty big gaps in total trips!

```python
%%bigquery --project dataviz-488717
SELECT
  start_station_name, COUNT(*)
AS total_trips
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
ORDER BY total_trips DESC;
```

5. Which Starting Point has the Most Trips?

This is a popular one, 21st/Speedway @ PCL with a whopping 108559 trips! I also included the stop with the least trips just out of curiosity. Only 1 trip in Eeyore's 2017! There are a couple others starting with only 1 point as well.

```python
%%bigquery --project dataviz-488717
SELECT
  start_station_name, COUNT(*)
AS total_trips
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
ORDER BY total_trips DESC
LIMIT 1;
```
