```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
y = iris.target

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Iris Sepal Length vs Width (Colored by Class)")
plt.show()
```

How do the colors work?
y contains 0, 1, 2 to represent the three iris classes in the dataset. "c=" is what gives each point a color based on the numbers 0, 1, 2. Then matplotlib assigns the default colors to the points on the graph. I don't remember how you weren't getting it in class.
