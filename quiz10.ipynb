Always start with the top five

```python
%%bigquery --project dataviz-488717

Select *
From `bigquery-public-data.austin_bikeshare.bikeshare_trips`
LIMIT 5
```

#### What was the second most popular starting station?
21st & Speedway @PCL with 71,145 trips! This is peculiar since the most trips in this data are 108,559 at 21st/Speedway @ PCL. A small punctuation change might be messing up the results a bit...

If that's the case, the next highest trips is 65,066 at Dean Keeton/Speedway.

```python
%%bigquery --project dataviz-488717
SELECT
    start_station_name,
    COUNT(*) AS total_trips
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
ORDER BY total_trips DESC
LIMIT 1 OFFSET 1;
```

#### How many trips lasted over an hour and were a round trip?

There were 70,666 round trips that were over an hour long. Admirable!

```python
%%bigquery --project dataviz-488717
SELECT
    COUNT(*) AS round_trips_over_one_hour
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE duration_minutes > 60
  AND start_station_name = end_station_name;
```
