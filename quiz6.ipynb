```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/Activity_Dataset_V1.csv"
df = pd.read_csv(url)

df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

sns.scatterplot(
    data=df,
    x="distance",
    y="calories",
    hue="workout_type",
    size="avg_speed",
    alpha=0.7
)

plt.title("Distance vs Calories (Workout Data)")
plt.show()
```
