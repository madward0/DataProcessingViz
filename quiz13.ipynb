```python
!pip install google-colab-selenium[undetected] > /dev/null

import google_colab_selenium as gs
import time
from selenium.webdriver.common.by import By

URL = "https://www.google.com"
TIMEOUT = 30

print("Initializing driver...")
driver = gs.UndetectedChrome()

try:
    print(f"Opening {URL}...")
    driver.get(URL)
    time.sleep(2)

    print("Extracting page info...")
    page_title = driver.title
    titles_list = [page_title] if page_title else []

    if titles_list:
        print(f"\nExtracted Titles:")
        for idx, title in enumerate(titles_list, 1):
            print(f"{idx}. {title}")
    else:
        print("No title found.")

    links = driver.find_elements(By.TAG_NAME, "a")
    if links:
        visible_links = [l for l in links if l.text.strip()]
        target = visible_links[0] if visible_links else links[0]
        print(f"\nFound links. Clicking: '{target.text}'...")
        driver.execute_script("arguments[0].click();", target)
        print("Click executed.")
    else:
        print("No links found.")

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    driver.quit()
    print("\nDriver closed.")
```
