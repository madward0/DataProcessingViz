2. Convert and find average GPA

3.7 avg GPA!

```python
data = {
    "age": ["20", "21", "19"],
    "gpa": ["3.5", "3.7", "3.9"],
    "enrolled": ["True", "False", "True"]
}

ages = [int(x) for x in data["age"]]
gpas = [float(x) for x in data["gpa"]]
enrolled = [x == "True" for x in data["enrolled"]]

avg_gpa = sum(gpas) / len(gpas)

print("Ages:", ages)
print("GPAs:", gpas)
print("Enrolled:", enrolled)
print("Average GPA:", avg_gpa)
```
