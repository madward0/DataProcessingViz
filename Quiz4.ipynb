```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/AB_NYC_2019.csv"
df = pd.read_csv(url)

max_prices = df.groupby('neighbourhood_group')['price'].max()

bars = plt.bar(max_prices.index, max_prices.values)

for bar in bars:
  height = bar.get_height()
  plt.text(bar.get_x() + bar.get_width()/2, height, f"${height:,.0f}", ha='center', va='bottom')

plt.title("Maximum Airbnb Price by Neighbourhood Group")
plt.xlabel("Neighbourhood Group")
plt.ylabel("Maximum Price ($)")
plt.show()
```

#### 2. Multiple Bar Graph

```python

avg_prices = df.groupby(['neighbourhood_group', 'room_type'])['price'].mean().unstack()

plt.figure
avg_prices.plot(kind='bar')

plt.title("Average Airbnb Price by Neighbourhood Group and Room Type")
plt.xlabel("Neighbourhood Group")
plt.ylabel("Average Price ($)")
plt.legend(title="Room Type")
plt.tight_layout()
plt.show()
```
