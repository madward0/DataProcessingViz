2b. Average trip miles based on payment type

The payment type with the most average miles is "Prcard" with 7.55 miles, and the least is "No Charge" at 2.02 miles. As for payed for trips, the lowest is "Pcard" at 2.44 miles.

```python
%%bigquery --project exam2-62322

SELECT payment_type,
  ROUND(AVG(trip_miles), 2)
  AS avg_trip_miles
FROM `exam2-62322.taxi_data.taxidata`
WHERE trip_miles IS NOT NULL
GROUP BY payment_type
ORDER BY avg_trip_miles DESC
```

2c. Fares above average

Among non-zero fares, there were 55515843 fares above the average fare. I also included the average fare out of curiosity, which is $31.87.

```python
%%bigquery --project exam2-62322

SELECT
  COUNT(*) AS num_above_avg,
  ROUND(AVG(fare), 2) AS avg_fare
FROM `exam2-62322.taxi_data.taxidata`
WHERE fare > 0
  AND fare > (
    SELECT AVG(fare)
    FROM `exam2-62322.taxi_data.taxidata`
    WHERE fare > 0
  )
```
