Import dataset and check null values. 20141 total!

```python
import pandas as pd

url = "https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/AB_NYC_2019.csv"
df = pd.read_csv(url)

df.head()
```

```python
print("Null values per column:\n")
print(df.isnull().sum())

print("\nTotal null values in dataset:", df.isnull().sum().sum())
```

Impute for null values

```python
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)

df['last_review'] = df['last_review'].fillna("No reviews")

df['host_name'] = df['host_name'].fillna("Unknown")

print("\nNull values after imputation:\n")
print(df.isnull().sum())
```
