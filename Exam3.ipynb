1a. Store as pandas df

```python
import pandas as pa
df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/refs/heads/main/Data_Sets/Activity_Dataset_V1.csv')

df.head(15)
```

1b. Remove "Unnamed"

```python
df = df.loc[:, ~df.columns.str.contains("^Unnamed")]
```

1c. Clean columns

```python
import re

clean_columns = []

for col in df.columns:
    col = re.sub(r"\(.*?\)", "", col)
    col = col.strip()
    clean_columns.append(col)

df.columns = clean_columns
df.head()
```

1d. Convert into datetime

```python
df["activity_day"] = pa.to_datetime(df["activity_day"])

print("\nUpdated Data Types:")
print(df.dtypes)
```

1e.

```python
df["total_steps"] = df["total_steps"].fillna(df["total_steps"].median())

df["total_steps"] = df["total_steps"].astype(int)

print(df["total_steps"].isna().sum())
```

1f.

```python
def convert_pace(pace):
    if pa.isna(pace) or pace == "":
        return None

    minutes, seconds = pace.split(":")
    return float(minutes) + float(seconds) / 60

df["avg_pace"] = df["avg_pace"].apply(convert_pace)

print(df["avg_pace"].head())
```

1g.

```python
calorie_stats = df.groupby("workout_type")["calories"].agg(
    ["mean", "median", "count", "std"]
)

print(calorie_stats)
```

1h.

```python
df["high_aerobic"] = df["aerobic"] >= 30

print(df[["aerobic", "high_aerobic"]].head())
```

1i.

```python
df["day_of_week"] = df["activity_day"].dt.day_name()

max_row = df.loc[df["max_cadence"].idxmax()]

print("Day of Week:", max_row["day_of_week"])
print("Workout Type:", max_row["workout_type"])
print("Max Cadence:", max_row["max_cadence"])
```
