```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Gather Iris Dataset

```python
df=pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/iris.csv')
df
```

```python
df.groupby(['Class'])['Class'].count()
```

### Bar Chart of Categorial Variable: Species/Class

```python
dfg = df.groupby(['Class'])['Class'].count()
dfg.plot.bar()
```

Well this bar chart sure is boring!

### More Visualizations: Sepal Length

```python
dfg=df.groupby(['SepalLength'])['SepalLength'].count()
dfg.plot.bar()
```

```python
df.SepalLength.value_counts().plot(kind='pie')
```

The pie chart is a little sloppy, but it is pretty! The bar chart worked best for this category so far.
### Histogram, Box Plot, and Violin

```python
df.SepalLength.plot.hist()
```

```python
df.SepalLength.plot(kind = 'box')
```

```python
sns.violinplot(data = df.SepalLength)
```

I've never seen a violin graph. It's cool looking!
